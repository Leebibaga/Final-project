```python
from scipy import stats
```

```python
variables = []
correlation_result = pd.DataFrame(index= data_num_columns.index, columns=["correlation"])
for key in data_num_columns.index:
    correlation, pvalue = stats.pearsonr(data[key], data["number_of_days_rented"])
    correlation_result.loc[key, "correlation"] = pvalue
    if pvalue < 0.05:
        variables.append(key)    
```

```python
pd.set_option("display.max_rows", None)
```

```python
for key in data_float_columns.index:
    correlation, pvalue = stats.pearsonr(data[key], data["number_of_days_rented"])
    if pvalue < 0.05:
        variables.append(key)   
```

```python
data_object_columns.index
```

```python
for key in data_object_columns.index: 
    unique_vals = data[key].unique()
    data[key].replace(to_replace=unique_vals, value= list(range(len(unique_vals))), inplace=True)
```

```python
data.head()
```

```python
for key in data_object_columns.index:    
    correlation, pvalue = stats.spearmanr(data["number_of_days_rented"], data[key])
    if pvalue < 0.05:
        variables.append(key)   
```

```python
### Variable Selection using Random Forest

from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

rfmod = RandomForestClassifier().fit(X, y)
#rfmod.feature_importances_ 

model = SelectFromModel(rfmod, prefit=True)
model.get_support()

varSel['RandomForest'] = model.get_support().astype('int64')
varSel
```

```python
import sklearn.metrics as metrics 

def regressionMetrics(y, yhat):
    res = {'MSE': metrics.mean_squared_error(y,yhat),
           'RMSE': np.sqrt(metrics.mean_squared_error(y,yhat)),
           'MAE': metrics.mean_absolute_error(y,yhat),
           'MSLE': metrics.mean_squared_log_error(y,yhat),
           'RMSLE': np.sqrt(metrics.mean_squared_log_error(y,yhat))
          }
    return res
```
